```python
#引数をデータとして差し替えて呼び出す関数
def generate_resipe_gemini(dish:str) -> str:
  prompt = f"以下の料理のレシピを考えてください　\n料理名: {dish}"
  model = genai.GenerativeModel("gemini-1.5-flash")
  response = model.generate_content(prompt)
  return response.text

#関数に引数を渡してメッセージ送信を実行
recipe_gemini = generate_resipe_gemini("カレー")
print(recipe_gemini)

recipe_gemini = generate_resipe_gemini("ラーメン")
print(recipe_gemini)
```

```python
def generate_spicy_recipe_gemini(dish: str) -> str :
  context = "前提条件:\n分量: 一人分\n味の好み: 激辛"
  prompt = f"前提条件を踏まえて、以下の料理のレシピを考えてください。\n{context}\n料理名:{dish}"
  model = genai.GenerativeModel('gemini-1.5-flash')
  response = model.generate_content(prompt)
  return response.text

#作った関数を呼び出す
response_str = generate_spicy_recipe_gemini("ラーメン")
print(response_str)
```

```python
import json
from IPython.display import JSON

def generate_json_recipe_gemini(dish:str) -> dict:
  prompt = f"""\
  ユーザーが入力した料理のレシピを考えてください。
  出力は以下のJSON形式にしてください。
  {{
    "材料": ["材料1","材料2"],
    "手順": ["手順1","手順2"]
  }}
  料理名: {dish}
  """
  model = genai.GenerativeModel('gemini-1.5-flash')
  response = model.generate_content(
    prompt,
    generation_config = {'response_mime_type':'application/json'}
  )
  return json.loads(response.text)

#関数を呼び出し
json_recipe_gemini = generate_json_recipe_gemini("カレー")
#JSON構造のデータを表示
display(JSON(json_recipe_gemini).data)
```
